```python
#n_hos_bed contain 498 row that means it has missing value
#fillna function is used to fill the rows which are missing
#here we are using mean of the n_hos_bed to fill the value
df.n_hos_beds = df.n_hos_beds.fillna(df.n_hos_beds.mean())
```

```python
df.info()
#now the n_hos_beds is 506
```

```python
#for filling all missing the value int the column at a time we can use
# df = df.fillna(df.mean)
#but this is not advisable as we might be filling some of the missing value with zero
#some with median and some with mode accordingly
```

```python
sns.jointplot(x="crime_rate", y="price", data = df)
```

```python
#the above curve looks like logarithimic curve it need to be transformed into the linear curve
#if we take log of crimerate then we can get kind of linear graph
#but since lots of value are near zero and log of zero is not defined and it tends towards -infinity to remove this
#we find log(x+1) where x is crime rate

df.crime_rate = np.log(1+df.crime_rate)
```

```python
sns.jointplot(x="crime_rate", y = "price", data= df)
```
